```python
#Import Dependencies
import matplotlib.pyplot as plt
```

```python
total_fare = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months,total_fare)
plt.xlabel('Month')
plt.ylabel('Savings $')
plt.title("Savings Account Results")
plt.show()
```

```python
# Create the plot with ax.plot()
fig, ax= plt.subplots()
ax.plot(months,total_fare)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(months, total_fare)
```

```python
plt.plot(months, total_fare, marker="D", color="green", linewidth=2, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.legend()
plt.grid()
```

```python
plt.bar(months, total_fare,color="green", label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
plt.barh(months, total_fare, color="magenta", label ='Boston')
plt.gca().invert_yaxis()
plt.ylabel("Months")
plt.xlabel("Total Fares")
plt.legend()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.scatter(x_axis, y_axis, c='red', label ='Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Months")
plt.legend()
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
fig,ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="skyblue", edgecolors="k", s=[i*5 for i in y_axis], label = "Boston")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
explode_values= (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, labels=x_axis, explode=explode_values, colors=colors, autopct="%.1f%%")
```
